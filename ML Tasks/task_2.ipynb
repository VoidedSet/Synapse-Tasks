### The previous commands were just to make sure we have a workable dataframe now, we'll do some manipulations on the data.

Find out the top 5 most voted candidates in Telangana

```python
df_telangana = df[df['State'] == 'Telangana'].sort_values(by='Total Votes', ascending=False)
df_telangana.drop(['ID'], axis=1).head(5).reset_index(drop=True)
```

Find out the total number of votes received by each state

```python
state_votes = df.groupby('State')['Total Votes'].sum().reset_index()
state_votes
```

Show these values as a percentage of the total number of votes received in the election.

```python
total_votes = state_votes['Total Votes'].sum()
state_votes['Percentage'] = (state_votes['Total Votes'] / total_votes) * 100

state_votes
```

Now, display the top 5 parties and find out the number of votes received by them

```python
state_votes.sort_values('Total Votes', ascending=False).head(5).reset_index(drop=True)
```

From the candidates who lost, list 10 candidates who got the highest percent votes.

```python
lossers = df[df['Result'] == 'Lost']
lossers.sort_values(by='% of Votes', ascending=False).head(10).reset_index(drop=True)
```

### These techniques help to understand the data that we have but we like something a bit more visual right? That's where matplotlib comes to our aid.

Make a piechart to show the votes recieved by the top 5 candidates in the elections.

```python
most_votes = df.nlargest(5, 'Total Votes')

plt.pie(most_votes['Total Votes'], labels=most_votes['Candidate'])
plt.title('Votes Received by Top 5 Candidates')
plt.show()
```

Make a bar graph to see the postal votes received by the parties in Karnataka

```python
post_votes = (df[df['State'] == 'Karnataka'].groupby('Party', as_index=False)['Postal Votes'].sum().nlargest(5, 'Postal Votes'))

plt.bar(post_votes['Party'], post_votes['Postal Votes'])
plt.title('Postal Votes Received by Parties in Karnataka')
plt.show()
```
